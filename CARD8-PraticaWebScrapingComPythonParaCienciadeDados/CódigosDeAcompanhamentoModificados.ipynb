# Web Scraping com Python para Ciência de Dados 

  Web Scraping consiste na coleta de dados a partir de um endereço web, como por exemplo, arquivos .html e .xml, e posterior análise e manipulação do mesmo

## Beautiful Soup
 Beautiful Soup é a principal biblioteca para ler e coletar dados de arquivos .html,seja local ou online, e com ele, é possível selecionar secções específicas utilizando as tags do html. É um pacote essêncial para Web Scraping, fornecendo saídas agradavéis e plena extração de dados.

```python
from bs4 import BeautifulSoup
```

### Uso do Web Scraping local

 Utilizando o BeautifulSoup para ler e extrair dados de arquivos .html locais 

```python
with open('home.html', 'r') as html_file: # Abre arquivo para leitura como arquivo html
    content = html_file.read()
    # print(content)
    
    soup = BeautifulSoup(content, 'lxml') # Atribui o conteudo do html em soup
    print(soup.prettify) # Printa o codigo mais organizado
    tags = soup.find_all('h5') # Atribuindo ao tags uma lista de todos os h5 que aparecem no código
    print(tags) # Printa a lista
```

### Metodos find() e find_all()

São metodos para localizar dados a partir das tags do arquivo html ou xml:

- find(): Localiza apenas um, é específico 

- find_all(): Localiza todos, com a tag desejada

```python
with open('home.html', 'r') as html_file: # Abre arquivo para leitura como arquivo html
    courses_html_tags = soup.find_all('h5') # Atribuindo ao tags uma lista de todos os h5 que aparecem no código
    
    # Printa separadamente cada h5
    for course in courses_html_tags:
        print(course.text)
```

### Utilizando o site para extrair dados
Para exercer o que foi aprendido em aula, utilizou-se todas as novas ferramentas aprendidas e outras aprendidas em outros cards. Foi criado um código para ler o top 100 artistas mais ouvidos do spotify, segundo o last.fm, onde é coletado seu nome, número de ouvintes e posição na lista, além da sua imagem de avatar do site e seu perfil. A aplicação também conta com a possibilidade do usuário escolher quantas posições ele deseja averiguar. As imagens coletadas são salvas na pasta de imagens e os dados de cada artisa é salvo em arquivos .txt na pasta artistas.
<br><br>Imagem do elemento correspondente a posição na lista de artistas mais ouvidos da semana na página web anteriormente aberta pelo requests. Aqui é feita a inspeção de elementos da página que são de interesse para a coleta de dados.<br> <br>

- Importando a biblioteca requests que foi anteriormente comentada


```python
import requests
```

- Utilizando requests, é atribuido as páginas 1 e 2 para as váriaveis ranksemanalpg1 e ranksemanalpg2, respectivamente, em formato de texto.

```python
ranksemanalpg1 =  requests.get('https://www.last.fm/pt/charts/weekly').text
ranksemanalpg2 =  requests.get('https://www.last.fm/pt/charts/weekly?page=2').text
print(ranksemanalpg1) 
print(ranksemanalpg2) 
```

Criação das principais funcionalidades do programa:
- Exibição de um rank cujo número de posições é definido pelo usuário
- Coleta do nome, número de ouvintes, posição e imagem do artista.
- Imagem é exibida e salva na pasta de imagens
- Os dados são salvos em um arquivo .txt
- É possível automatizar o programa para que ele rode a cada 10 minutos

```python
%matplotlib inline 
import matplotlib.image as mpimg
import matplotlib.pyplot as plt

import time 


soup = BeautifulSoup(ranksemanalpg1, 'lxml')
soup2 = BeautifulSoup(ranksemanalpg2, 'lxml')

#print(ranksemanalpg1)

# Utilizando soup para encontrar todos os itens da lista de artistas da primeira página e segunda página.
artistas = soup.find_all('tr', class_ = 'js-link-block weeklychart-item') 
artistas2 = soup2.find_all('tr', class_ = 'js-link-block weeklychart-item')
 
# Requisitando o número de posições ao usuário utilizando input.

print('Deseja vizualizar quantos artistas?')
artistainput = input('>')
print(f'top {artistainput} artistas semanais')

# função para listar os artistas e suas caracteristicas
#def find_artista():

# for para percorrer todos os artistas da página, enumerados por indíces(index), que serão utilizados para a criação de arquivos.
for index, artista in enumerate(artistas):
    
    # Utilizando find para localizar a posição, nome e ouvintes do artista, como texto sem espaços, a partir de sua tag e classe, que sõo fixas para todas a posições
    posicao_do_artista = artista.find('td', 'weeklychart-rank').text.replace(' ', '')
    nome_do_artista = artista.find('a', class_='link-block-target').text.replace(' ', '')
    ouvintes_do_artista = artista.find('td', class_ = 'weeklychart-listeners').text.replace(' ', '')

    # Coletando url da imagem.
    imagem_artistaurl = artista.find('img').attrs['src']
    imagem_artista = requests.get(imagem_artistaurl).content

    # Definindo o nome do arquivo.
    nome_imagem = 'imagens/a' + str(index) + '.jpg'
    
    # Salvando a imagem pelo url e com o nome definido.
    with open(nome_imagem, 'wb') as handler:
        handler.write(imagem_artista) # Criação do arquivo pelo url da imagem
    
    # Atribuindo a url da página do artista no last.fm, para que o usuário possa acessar mais informações.    
    maisinfo = 'https://www.last.fm'+ artista.find('a', class_='link-block-target')['href']
    
    # Exibindo as informações do artista, com o metodo strip para eliminar espaçoes desnecessários
    print(f"Posição: {posicao_do_artista.strip()}")
    print(f"Artista: {nome_do_artista.strip()}")
    print(f"Número de ouvintes: {ouvintes_do_artista.strip()}")
    print(f"Mais informações sobre o artista: {maisinfo}")
    print('Imagem: ')
    
    # Exibindo a imagem utilizando matplotlib
    imagem_artista = mpimg.imread(nome_imagem) # leitura da imagem
    imgplot = plt.imshow(imagem_artista) # Abertura da imagem
    plt.axis('off') # Remove os eixos x e y
    plt.show() # Exibição da imagem
    
    # Armazenamento dos dados em um arquivo .txt
    with open (f'artistas/a{index}.txt', 'w') as f: 
        
        # Escrevendo com o metodo strip para eliminar espaçoes desnecessários
        f.write(f"Posição: {posicao_do_artista.strip()}\n")
        f.write(f"Artista: {nome_do_artista.strip()}\n")
        f.write(f"Número de ouvintes: {ouvintes_do_artista.strip()}\n")
    
    # If para sair do loop se atingir a posição passada pelo usuário
    if(int(posicao_do_artista) == int(artistainput)):
        break

# If para acessar a segunda página somente se for necessário, se passar de 50 artistas        
if int(artistainput)>50:
    # for para percorrer todos os artistas da segunda página, enumerados por indíces(index), que serão utilizados para a criação de arquivos.
    for index, artista in enumerate(artistas2):
        
        # Utilizando find para localizar a posição, nome e ouvintes do artista, como texto sem espaços, a partir de sua tag e classe, que sõo fixas para todas a posições
        posicao_do_artista = artista.find('td', 'weeklychart-rank').text.replace(' ', '')
        nome_do_artista = artista.find('a', class_='link-block-target').text.replace(' ', '')
        ouvintes_do_artista = artista.find('td', class_ = 'weeklychart-listeners').text.replace(' ', '')

        # Coletando url da imagem e seu contéudo.
        imagem_artistaurl = artista.find('img').attrs['src']
        imagem_artista = requests.get(imagem_artistaurl).content
        
        # Definindo o nome do arquivo.
        nome_imagem = 'imagens/a' + str(index) + '.jpg'
        
        # Salvando a imagem pelo url e com o nome definido.
        with open(nome_imagem, 'wb') as handler:
            handler.write(imagem_artista) # Criação do arquivo pelo url da imagem
            
        # Atribuindo a url da página do artista no last.fm, para que o usuário possa acessar mais informações.  
        maisinfo = 'https://www.last.fm'+ artista.find('a', class_='link-block-target')['href']
        
        # Exibindo as informações do artista, com o metodo strip para eliminar espaçoes desnecessários
        print(f"Posição: {posicao_do_artista.strip()}")
        print(f"Artista: {nome_do_artista.strip()}")
        print(f"Número de ouvintes: {ouvintes_do_artista.strip()}")
        print(f"Mais informações sobre o artista: {maisinfo}")
        print('Imagem: ')
                
        # Exibindo a imagem utilizando matplotlib
        imagem_artista = mpimg.imread(nome_imagem) # leitura da imagem
        imgplot = plt.imshow(imagem_artista) # Abertura da imagem
        plt.axis('off') # Remove os eixos x e y
        plt.show() # Exibição da imagem
            
        # Armazenamento dos dados em um arquivo .txt
        with open (f'artistas/a{index}.txt', 'w') as f:
            
            # Escrevendo com o metodo strip para eliminar espaçoes desnecessários
            f.write(f"Posição: {posicao_do_artista.strip()}\n")
            f.write(f"Artista: {nome_do_artista.strip()}\n")
            f.write(f"Número de ouvintes: {ouvintes_do_artista.strip()}")
        
        # If para sair do loop se atingir a posição passada pelo usuário
        if(int(posicao_do_artista) == int(artistainput)):
            break
# If para a criação da automatização, se o programa for executado na main.
#if __name__ == '__main__':
#   # Enquanto ele estiver sendo executado... 
#   while True:
#        find_artista() # Função para coletar e exivir os artistas
#        tempo_de_espera = 10 # Intervalo de execução
#        print(f'Esperando {tempo_de_espera} minutos...')
#        time.sleep(tempo_de_espera + 60) # Definindo a espera pelo intervalo de execução
```
