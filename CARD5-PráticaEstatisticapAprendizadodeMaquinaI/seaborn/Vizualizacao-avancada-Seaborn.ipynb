# Seaborn: Bibioteca de visualização

é baseada em matplotlib, extremamente importante para vizualização avançada de dados, com gráficos detalhados e bonitos. Além de visual, também é adicionado novas funcionalidades para interpretar dados.

> A partir de dados csv de um site da Web sobre eficiência de motor de carro, é criado um dataframe e utilizado para criar gráficos

```python
%matplotlib inline

import pandas as pd

df = pd.read_csv("http://media.sundog-soft.com/SelfDriving/FuelEfficiency.csv")

gear_counts = df['# Gears'].value_counts()

gear_counts.plot(kind='bar') # gráfico de barras sobre a quantidade de marchas
```

```python
import seaborn as sns
sns.set() # habilitando configurações mais agradáveis com seaborn
```

```python
gear_counts.plot(kind='bar') # Exibindo o mesmo gráfico, porém utilizando seaborn
```

```python
df.head() # Mostrando o DataFrame
```

```python
sns.displot(df['CombMPG']) # Histograma 
```

```python
df2 = df[['Cylinders', 'CityMPG', 'HwyMPG', 'CombMPG']] # Criando DataFrame com as colunas especificadas
df2.head()
```

```python
sns.pairplot(df2, height=2.5);  # Pair Plot, combina todos os plots com os atributos, com o df2
```

```python
sns.scatterplot(x="Eng Displ", y="CombMPG", data=df) # Graáfico de dispersão
```

```python
sns.jointplot(x="Eng Displ", y="CombMPG", data=df) # Combina scatterplot com o histograma em seus eixos
```

```python
sns.lmplot(x="Eng Displ", y="CombMPG", data=df) # Gráfico de dispersão com uma linha de regressão linear
```

```python
# Crianado um gráfico de caixa com os nomes dos pontos do eixo x rotacionados em 45 graus
sns.set(rc={'figure.figsize':(15,5)})
ax=sns.boxplot(x='Mfr Name', y='CombMPG', data=df)
ax.set_xticklabels(ax.get_xticklabels(),rotation=45)
```

```python
# Pontos de dados individuais agrupados
ax=sns.swarmplot(x='Mfr Name', y='CombMPG', data=df)
ax.set_xticklabels(ax.get_xticklabels(),rotation=45)
```

```python
# Um histograma para dados categóricos do mfr names
ax=sns.countplot(x='Mfr Name', data=df)
ax.set_xticklabels(ax.get_xticklabels(),rotation=45)
```

```python
# Gráfico em forma de mapa de calor 
df2 = df.pivot_table(index='Cylinders', columns='Eng Displ', values='CombMPG', aggfunc='mean') # Novo DataFrame reorganizando o original
sns.heatmap(df2)
```
