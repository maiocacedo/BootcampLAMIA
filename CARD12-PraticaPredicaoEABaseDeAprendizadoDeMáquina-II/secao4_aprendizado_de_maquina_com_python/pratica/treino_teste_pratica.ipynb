## Treino / Teste
 _Prática da aula sobre **Treino e teste** em modelos de predição com regressão polinomial sobre a **relação** entre **velocidade da página** e **quantidade de produtos comprados** com base no que foi visto em aula_

Testando diferentes parâmetros para tentar melhorar os resultados, com a mesmo dataset.

```python
%matplotlib inline
import numpy as np
from pylab import *
import matplotlib.pyplot as plt

np.random.seed(2) 

pageSpeeds = np.random.normal(3.0, 1.0, 100) # Dados de velocidade de página
purchaseAmount = np.random.normal(50.0, 30.0, 100) / pageSpeeds # Dados de valor de compra 


scatter(pageSpeeds, purchaseAmount) # Gráfico de dispersão
plt.show()
```

Dividindo os dados: 80% para treinamento e 20% para teste

```python
trainX = pageSpeeds[:80] # Dados de treinamento
testX = pageSpeeds[80:] # Dados de teste

trainY = purchaseAmount[:80] # Dados de treinamento
testY = purchaseAmount[80:] # Dados de teste
```

Gráficos de dispersão dos dados

```python
scatter(trainX, trainY, c='b') # Gráfico de dispersão dos dados de treinamento
plt.show()
```

```python
scatter(testX, testY, c='purple') # Gráfico de dispersão dos dados de teste
plt.show()
```

Aplicando um polinômio de grau **5** nos dados de treino

```python
x = np.array(trainX) # Dados de treinamento
y = np.array(trainY) # Dados de treinamento

p4 = np.poly1d(np.polyfit(x, y, 5)) # Ajuste de polinômio de grau 5
```

```python
import matplotlib.pyplot as plt

# Gráfico de dispersão dos dados de treinamento
xp = np.linspace(0, 7, 100)
axes = plt.axes()
axes.set_xlim([0,7])
axes.set_ylim([0, 200])
plt.scatter(x, y, s=10, c='b', label='Dados de Treinamento')
plt.plot(xp, p4(xp), c='r', label='Ajuste Polinomial de Grau 5')
plt.title('Ajuste Polinomial de Grau 5')
plt.show()
```

Agora, aplicando nos dados de teste


```python
testx = np.array(testX) # Dados de teste
testy = np.array(testY) # Dados de teste

# Gráfico de dispersão dos dados de teste
axes = plt.axes()
axes.set_xlim([0,7])
axes.set_ylim([0, 200])
plt.scatter(testx, testy, c='purple', label='Dados de Teste')
plt.plot(xp, p4(xp), c='r', label='Ajuste Polinomial de Grau 5')
plt.show()
```

Pessimo $ R^2 $.

```python
from sklearn.metrics import r2_score

r2 = r2_score(testy, p4(testx)) # Cálculo do R²

print(r2)
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(np.array(trainY), p4(np.array(trainX))) # Cálculo do R² para os dados de treinamento

print(r2)
```
