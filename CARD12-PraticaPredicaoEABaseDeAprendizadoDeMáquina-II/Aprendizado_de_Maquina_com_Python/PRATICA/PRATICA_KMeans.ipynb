## K-Means Clustering
 _Prática da aula sobre  **K-Means Clustering** Aplicado a dados ficticios relacionando idade e receita, com base no que foi visto em aula_

Fabricando função para gerar dados falsos

```python
from numpy import random, array

# função para criar dados de entrada para o algoritmo K-means: receita e idade
# O número de pontos de dados é N e o número de clusters é k
def createClusteredData(N, k):
    random.seed(10)
    pointsPerCluster = float(N)/k
    X = []
    for i in range (k):
        incomeCentroid = random.uniform(20000.0, 200000.0) # receita média do cluster
        ageCentroid = random.uniform(20.0, 70.0) # idade média do cluster
        # gera pontos de dados aleatórios em torno do centróide
        for j in range(int(pointsPerCluster)):
            X.append([random.normal(incomeCentroid, 10000.0), random.normal(ageCentroid, 2.0)])
    X = array(X) # converte a lista de pontos de dados em um array numpy
    return X
```

Aplicando K-Means Clustering

```python
%matplotlib inline

from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale
from numpy import random

# Criando dados de entrada para o algoritmo K-means, com a função: receita e idade
data = createClusteredData(100, 5)

model = KMeans(n_clusters=5) # Definindo o número de clusters como 5

# Ajustando o modelo aos dados
model = model.fit(scale(data))

# Exibindo os centros dos clusters
print(model.labels_)
plt.figure(figsize=(8, 6))
plt.scatter(x=data[:,0], y=data[:,1], c=model.labels_.astype(float))
plt.show()
```
