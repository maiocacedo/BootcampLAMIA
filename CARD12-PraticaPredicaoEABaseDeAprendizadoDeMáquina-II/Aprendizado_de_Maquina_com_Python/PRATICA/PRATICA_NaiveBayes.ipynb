## Naive Bayes 
 _Prática da aula sobre  **Naive Bayes** em um conjunto de dados referente a **emails**, utilizado para detectar spams, com base no que foi visto em aula. Como extra, apliquei o método train/test._

Criando dateframe dos arquivos de emails e aplicando Train/Test

```python
import os
import io
import numpy
import pandas as pd
from pandas import DataFrame
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

# Função para ler arquivos
def readFiles(path):
    for root, dirnames, filenames in os.walk(path):
        for filename in filenames:
            path = os.path.join(root, filename)

            inBody = False
            lines = []
            f = io.open(path, 'r', encoding='latin1')
            for line in f:
                if inBody:
                    lines.append(line)
                elif line == '\n':
                    inBody = True
            f.close()
            message = '\n'.join(lines)
            yield path, message

# Função para criar um DataFrame a partir de arquivos em um diretório
def dataFrameFromDirectory(path, classification):
    rows = []
    index = []
    for filename, message in readFiles(path):
        rows.append({'message': message, 'class': classification})
        index.append(filename)

    return DataFrame(rows, index=index)

# Criando DataFrames para spam e ham
data = DataFrame({'message': [], 'class': []})

dataSpam = pd.concat([data, dataFrameFromDirectory("emails/spam", "spam")]);

dataHam = pd.concat([data, dataFrameFromDirectory("emails/ham", "ham")])

allData = pd.concat([dataSpam, dataHam])

# Aplicando o Train/Test Split
# 80% para treino e 20% para teste

dataSpamTrain = dataSpam.sample(frac=0.8, random_state=1) # random_state=1 para reprodutibilidade frac=0.8 significa 80% do dataset
dataSpamTest = dataSpam.drop(dataSpamTrain.index) # drop() remove os dados que estão no train set do test set

dataHamTrain = dataHam.sample(frac=0.8, random_state=1) # random_state=1 para reprodutibilidade frac=0.8 significa 80% do dataset
dataHamTest = dataHam.drop(dataHamTrain.index) # drop() remove os dados que estão no train set do test set

dataTrain = pd.concat([dataSpamTrain, dataHamTrain]) # concat() junta os dois dataframes, o de spam e o de ham, em um só dataframe de treino
dataTest = pd.concat([dataSpamTest, dataHamTest]) # concat() junta os dois dataframes, o de spam e o de ham, em um só dataframe de teste

```

Separando uma lista de palavras e aplicando o Naive Bayes Multinomial

```python
vectorizer = CountVectorizer() # Vetorizador
counts = vectorizer.fit_transform(dataTrain['message'].values) # fit_transform() aplica o vetorizer nos dados de treino, transformando-os em vetores numéricos

classifier = MultinomialNB() # Classificador Naive Bayes Multinomial
targets = dataTrain['class'].values # targets são os rótulos (spam ou ham) dos dados de treino
classifier.fit(counts, targets) # fit() aplica o classificador nos dados de treino, ajustando os parâmetros do classificador para os dados de treino
```

Testando


```python
predictions = classifier.predict(vectorizer.transform(dataTest['message'].values)) # transform() aplica o vetorizer nos dados de teste, transformando-os em vetores numéricos
predictions 
```
