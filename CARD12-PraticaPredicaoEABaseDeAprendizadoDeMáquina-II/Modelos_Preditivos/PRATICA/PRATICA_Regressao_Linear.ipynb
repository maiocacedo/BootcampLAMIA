## Regressão Linear - Prática 
 _Praticando a **regressão linear** em dados que mostram a **relação linear** entre **velocidade da página** e **quantidade de produtos comprados** com base no que foi visto em aula_

Gerando os dados:

```python
import numpy as np
import matplotlib.pyplot as plt

velocidadePagina = np.random.normal(3.0,0.8,1000) # velocidade média é 3 com desvio padrão igual a 0.8 em 1000 amostras
quantidadeCompra = 100 - (velocidadePagina + np.random.normal(0,0.2,1000)) * 3 # quanto mais rápido, mais compras

plt.grid()
plt.title("Velocidade da página vs Compras")
plt.xlabel("Velocidade da página")
plt.ylabel("Quantidade de compras")
plt.scatter(velocidadePagina, quantidadeCompra) # gráfico de dispersão
```

Criando o modelo de regressão linear com as duas características relacionadas entre si, velocidade da página e quantidade de compras, utilizando numpy

```python
from scipy import stats

slope, intercept, r_value, p_value, std_err = stats.linregress(velocidadePagina, quantidadeCompra) # regressão linear
```

Ótimo coeficiente de determinação 

```python
r_value ** 2 # coeficiente de determinação (r-squared - R²) - quanto mais próximo de 1, melhor a relação entre as variáveis e mais próximo de 0, pior a relação entre as variáveis
```

Utilizando slope e intercept para plotar valores estimados vs. observados

```python
import matplotlib.pyplot as plt

def predicao(x):
    return slope * x + intercept # função de predição

fitLine = predicao(velocidadePagina) # linha de predição

plt.scatter(velocidadePagina, quantidadeCompra) # gráfico de dispersão
plt.plot(velocidadePagina, fitLine, c='r') # linha de predição
plt.xlabel("Velocidade da página")
plt.ylabel("Quantidade de compras")
plt.show() # mostra o gráfico
```
