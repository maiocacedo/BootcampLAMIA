## Regressão polinomial
 _Acompanhamento da aula sobre  **regressão polinomial** em dados que mostram a **relação** entre **velocidade da página** e **quantidade de produtos comprados**_

```python
# importando as bibliotecas necessárias
import numpy as np 
import matplotlib.pyplot as plt

np.random.seed(2) # para reprodutibilidade

# gerando dados aleatórios para pageSpeeds e purchaseAmount
pageSpeeds = np.random.normal(3.0, 1.0, 1000)
purchaseAmount = np.random.normal(50.0,10.0,1000) / pageSpeeds

# grafico de dispersão
plt.scatter(pageSpeeds, purchaseAmount)
```

Criando nosso polinômio

```python
x = np.array(pageSpeeds) # converte pageSpeeds para um array numpy
y = np.array(purchaseAmount) # converte purchaseAmount para um array numpy

p4 = np.poly1d(np.polyfit(x, y, 4)) # ajuste de polinômio de grau 4
```

Criando um gráfico com nosso polinômio com x variando de 0 a 7

```python
xp = np.linspace(0, 7, 100) # cria um array de 100 pontos entre 0 e 7 para x
plt.scatter(x, y, c='blue', s=1) # plota os dados originais
plt.plot(xp, p4(xp), c='red') # plota a curva ajustada
plt.show() # exibe o gráfico
```

Calculando o r quadratico

```python
from sklearn.metrics import r2_score

# Calculando a métrica r2, que mede a qualidade do ajuste
r2 = r2_score(y, p4(x))
print(f"R^2: {r2:f}") 
```
