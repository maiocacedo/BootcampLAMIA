## Regressão polinomial
 _Prática da aula sobre  **regressão polinomial** em dados que mostram a **relação** entre **velocidade da página** e **quantidade de produtos comprados** com base no que foi visto em aula_

Para praticar, comparei os valores entre os graus de polinomio da regressão, utilizando os mesmos dados da aula.

```python
# importando as bibliotecas necessárias
import numpy as np 
import matplotlib.pyplot as plt

np.random.seed(2) # para reprodutibilidade

# gerando dados aleatórios para pageSpeeds e purchaseAmount
pageSpeeds = np.random.normal(3.0, 1.0, 1000)
purchaseAmount = np.random.normal(50.0,10.0,1000) / pageSpeeds

# grafico de dispersão
plt.scatter(pageSpeeds, purchaseAmount, s=3, alpha=0.5, c='purple')
```

Criando nosso polinômio

```python
x = np.array(pageSpeeds) # converte pageSpeeds para um array numpy
y = np.array(purchaseAmount) # converte purchaseAmount para um array numpy

p4 = np.poly1d(np.polyfit(x, y, 4)) # ajuste de polinômio de grau 4
```

Criando um gráfico com nosso polinômio com x variando de 0 a 7

```python
xp = np.linspace(0, 7, 100) # cria um array de 100 pontos entre 0 e 7 para x
plt.scatter(pageSpeeds, purchaseAmount, s=3, alpha=0.5, c='purple') # plota os dados originais
plt.plot(xp, p4(xp), c='red') # plota a curva ajustada
plt.show() # exibe o gráfico
```

Calculando o r quadratico

```python
from sklearn.metrics import r2_score

# Calculando a métrica r2, que mede a qualidade do ajuste
r2p4 = r2_score(y, p4(x))
print(f"R^2: {r2p4:f}") 
```

Repetindo o processo, mas para grau 5

```python
x = np.array(pageSpeeds) # converte pageSpeeds para um array numpy
y = np.array(purchaseAmount) # converte purchaseAmount para um array numpy

p5 = np.poly1d(np.polyfit(x, y, 5)) # ajuste de polinômio de grau 5
```

```python
xp = np.linspace(0, 7, 100) # cria um array de 100 pontos entre 0 e 7 para x
plt.scatter(pageSpeeds, purchaseAmount, s=3, alpha=0.5, c='purple') # plota os dados originais
plt.plot(xp, p5(xp), c='red') # plota a curva ajustada
plt.show() # exibe o gráfico
```

```python
# Calculando a métrica r2, que mede a qualidade do ajuste
r2p5 = r2_score(y, p5(x))
print(f"R^2: {r2p5:f}") 
```

Repetindo o processo, mas para grau 3

```python
x = np.array(pageSpeeds) # converte pageSpeeds para um array numpy
y = np.array(purchaseAmount) # converte purchaseAmount para um array numpy

p3 = np.poly1d(np.polyfit(x, y, 3)) # ajuste de polinômio de grau 3
```

```python
xp = np.linspace(0, 7, 100) # cria um array de 100 pontos entre 0 e 7 para x
plt.scatter(pageSpeeds, purchaseAmount, s=3, alpha=0.5, c='purple') # plota os dados originais
plt.plot(xp, p3(xp), c='red') # plota a curva ajustada
plt.show() # exibe o gráfico
```

```python
# Calculando a métrica r2, que mede a qualidade do ajuste
r2p3 = r2_score(y, p3(x))
print(f"R^2: {r2p3:f}") 
```

Comparando os r2_score:

```python
print(f"R^2: {r2p3:f} (grau 3)")
print(f"R^2: {r2p4:f} (grau 4)")
print(f"R^2: {r2p5:f} (grau 5)")
```

O de grau 3 é o pior;
O de grau 4 é bom;
E o de grau 5 é o melhor e não aparenta estar em overfitting.
